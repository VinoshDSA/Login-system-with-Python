```python
import re
regex = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'
reg1 = "^(?=.*[a-z])(?=.*[A-Z])(?=.*\d)(?=.*[@$!%*#?&])[A-Za-z\d@$!#%*?&]{5,16}$"

granted=False
def grant():
    global granted
    granted=True
def login(uname,passd):
    success=False
    file=open("task1.txt","r")
    d=[]
    f=[]
    
    for i in file:
        a,b=i.split(",")
        b=b.strip()
        a=a.strip()
        d.append(a)
        f.append(b)
        data=dict(zip(d,f))
        
        
        if (a==uname and b==passd):
            success=True
            break
    print(data)
    file.close()
    if(success):
        print("logged in successfully!!!")
        grant()
    else:
        print("Username or password not matched..pls signin")
def register(uname,passd):
    print("Registered successfully")
    file=open("task1.txt","a")
    file.write(uname + "," + passd + "\n")
    
    file.close()
def access(option):
    global uname
    if option =="Login":
        uname=input("Enter username: ")
        passd = input("Enter Password: ")
        login(uname,passd)
    else:
        print("Please Signin:")
        uname=input("Create username(email): ")
        vali=False
        if(re.fullmatch(regex, uname)):
            #print("Valid Email")
            vali=True
            
            if vali:
                passd = input("Create Password: ")
                pat = re.compile(reg1)
                mat = re.search(pat, passd)
                if mat:
                    passd1=input("Confirm password: ")
                    if passd==passd1:
                        register(uname,passd)
                        
                    else:
                        print("Create Password and confirm password not matched")
                        home()
                    
                else:
                    print("Invalid password")
                    home()
                    
                #register(uname,passd)
                
        else:
            print("Invalid Email")
            home()
            
def forgtpassd():
    file=open("task1.txt","r")
    d=[]
    f=[]
    
    for i in file:
        a,b=i.split(",")
        b=b.strip()
        a=a.strip()
        d.append(a)
        f.append(b)
        data=dict(zip(d,f))
    print(data)
    asd=input("Pls enter the Username:")
    try:
        for i in data.keys():
            #print(i)
            if i==asd:
                usernaame=data[i]
        print("Your password is: ",usernaame)
 
    except:
        print("Username not exist..please sign in")
        home()
    
def home():
    global option
    print("Welcome to Guvi Bootcamp")
    option =input("Login | Reg | forgtpassd: ")
    
    if (option == "Login"):
        access(option)
    elif(option =="Reg"):
        access(option)
    elif(option =="forgtpassd"):
        forgtpassd()
    else:
        home()


home()
#access(option)

if(granted):
    print("Welcome to Guvi Bootcamp")
    print("### USER DETAILS ###")
    print("Username: ",uname)
```
